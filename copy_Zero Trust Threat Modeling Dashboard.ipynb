## ‚úÖ `src/data_gen/generate_assets.py`

```python
import pandas as pd
import random
from faker import Faker
from pathlib import Path

fake = Faker()

def generate_assets(num_assets=50):
    asset_types = ['Workstation', 'Server', 'Cloud VM', 'Mobile', 'Database']
    locations = ['On-Prem', 'AWS', 'Azure', 'GCP']
    sensitivity_levels = ['High', 'Medium', 'Low']
    tag_pool = ['finance', 'devops', 'hr', 'customer-data', 'logs', 'security']

    assets = []
    for i in range(1, num_assets + 1):
        asset_id = f"A-{i:03d}"
        asset_type = random.choice(asset_types)
        owner = fake.email()
        location = random.choice(locations)
        sensitivity = random.choices(sensitivity_levels, weights=[0.3, 0.5, 0.2])[0]
        tags = ", ".join(random.sample(tag_pool, k=random.randint(1, 3)))

        assets.append({
            "asset_id": asset_id,
            "asset_type": asset_type,
            "owner": owner,
            "location": location,
            "sensitivity": sensitivity,
            "tags": tags
        })

    df = pd.DataFrame(assets)
    return df

if __name__ == "__main__":
    output_path = Path(__file__).resolve().parents[2] / "data" / "assets.csv"
    df_assets = generate_assets()
    df_assets.to_csv(output_path, index=False)
    print(f"‚úÖ Generated {len(df_assets)} assets to {output_path}")
```

````python
import pandas as pd
import random
from faker import Faker
from datetime import datetime, timedelta
import os

# Initialize Faker
fake = Faker()
Faker.seed(0)

# Config
num_rows = 5000
users = [f"u{str(i).zfill(3)}" for i in range(1, 101)]
roles = ['Admin', 'Employee', 'Contractor', 'Guest']
resources = ['/admin/panel', '/files/projectX.docx', '/email/inbox', '/vpn/connect', '/dev/api']
actions = ['login', 'logout', 'download', 'upload', 'delete']
locations = ['US', 'UK', 'DE', 'IN', 'CN']

def generate_log():
    timestamp = fake.date_time_between(start_date='-30d', end_date='now')
    user = random.choice(users)
    role = random.choice(roles)
    resource = random.choice(resources)
    action = random.choice(actions)
    location = random.choice(locations)
    mfa = random.choice([True, False])
    success = random.choice([True, False])
    risk_score = round(random.uniform(0.0, 1.0), 2)
    
    return {
        "Timestamp": timestamp,
        "UserID": user,
        "Role": role,
        "Resource Accessed": resource,
        "Action": action,
        "Location": location,
        "MFA Used": mfa,
        "Success": success,
        "Risk Score": risk_score
    }

# Generate data
logs = [generate_log() for _ in range(num_rows)]
df = pd.DataFrame(logs)

# Save to CSV
data_path = r"C:\Users\acer\ZeroTrustThreatModelingDashboard\data\access_logs.csv"
df.to_csv(data_path, index=False)

print(f"Fake access log data saved to:\n{data_path}")
df.head()
````

````python
# Generate Identity Directory Dataset
identity_data = []

for user in users:
    name = fake.name()
    email = f"{user}@zerotrustdemo.com"
    department = random.choice(['Engineering', 'HR', 'Finance', 'Security', 'Marketing'])
    role = random.choice(roles)
    join_date = fake.date_between(start_date='-5y', end_date='-1y')
    pwd_change = fake.date_between(start_date=join_date, end_date='today')
    
    identity_data.append({
        "UserID": user,
        "Name": name,
        "Email": email,
        "Department": department,
        "Role": role,
        "Join Date": join_date,
        "Last Password Change": pwd_change
    })

identity_df = pd.DataFrame(identity_data)
identity_df.to_csv(r"C:\Users\acer\ZeroTrustThreatModelingDashboard\data\identity_directory.csv", index=False)

print("Identity directory saved.")
identity_df.head()
````

````python
# Generate Security Alerts
alert_types = ['Brute Force', 'Phishing', 'Privilege Escalation', 'Unusual Access Time', 'Data Exfiltration']
severity_levels = ['Low', 'Medium', 'High', 'Critical']

alerts = []
for i in range(1, 1001):
    timestamp = fake.date_time_between(start_date='-30d', end_date='now')
    user = random.choice(users)
    alert_type = random.choice(alert_types)
    severity = random.choices(severity_levels, weights=[0.4, 0.3, 0.2, 0.1])[0]
    description = fake.sentence(nb_words=6)
    resolved = random.choice([True, False])
    
    alerts.append({
        "AlertID": f"A{i:04}",
        "Timestamp": timestamp,
        "UserID": user,
        "Alert Type": alert_type,
        "Severity": severity,
        "Description": description,
        "Resolved": resolved
    })

alerts_df = pd.DataFrame(alerts)
alerts_df.to_csv(r"C:\Users\acer\ZeroTrustThreatModelingDashboard\data\security_alerts.csv", index=False)

print("Security alerts dataset saved.")
alerts_df.head()
````

````python
# Generate Network Events
protocols = ['TCP', 'UDP', 'ICMP', 'HTTP', 'HTTPS']
actions = ['Allow', 'Block', 'Reset']

network_events = []
for i in range(1, 2001):
    timestamp = fake.date_time_between(start_date='-30d', end_date='now')
    source_ip = fake.ipv4_public()
    destination_ip = fake.ipv4_public()
    protocol = random.choice(protocols)
    action = random.choice(actions)
    bytes_sent = random.randint(100, 50000)
    risk_score = round(random.uniform(0, 1), 2)

    network_events.append({
        "EventID": f"N{i:05}",
        "Timestamp": timestamp,
        "Source IP": source_ip,
        "Destination IP": destination_ip,
        "Protocol": protocol,
        "Action": action,
        "Bytes Sent": bytes_sent,
        "Risk Score": risk_score
    })

network_df = pd.DataFrame(network_events)
network_df.to_csv(r"C:\Users\acer\ZeroTrustThreatModelingDashboard\data\network_events.csv", index=False)

print("Network events dataset saved.")
network_df.head()
````

````python
# Load datasets
access_df = pd.read_csv(r"C:\Users\acer\ZeroTrustThreatModelingDashboard\data\access_logs.csv", parse_dates=['Timestamp'])
alerts_df = pd.read_csv(r"C:\Users\acer\ZeroTrustThreatModelingDashboard\data\security_alerts.csv", parse_dates=['Timestamp'])

# Filter failed logins
failed_df = access_df[access_df['Success'] == False]

# Count failed logins per user
failed_counts = failed_df.groupby('UserID').size().reset_index(name='Failed Logins')

# Count alerts per user
alert_counts = alerts_df.groupby('UserID').size().reset_index(name='Security Alerts')
````

````python
# Merge the two summaries
user_risk = pd.merge(failed_counts, alert_counts, on='UserID', how='outer').fillna(0)

# Convert to int
user_risk['Failed Logins'] = user_risk['Failed Logins'].astype(int)
user_risk['Security Alerts'] = user_risk['Security Alerts'].astype(int)

# Add a combined risk score
user_risk['Total Risk Events'] = user_risk['Failed Logins'] + user_risk['Security Alerts']

# Sort by total risk
risky_users = user_risk.sort_values(by='Total Risk Events', ascending=False)

# Display top 10
risky_users.head(10)
````
